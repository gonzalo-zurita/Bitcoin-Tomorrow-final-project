```python
fig = plt.figure(figsize= (12,8))
plot1 = sns.lineplot(x='time', y='close_btc', data= features, color='blue')
plot2 = sns.lineplot(x='time', y='close_gold', data= features, color='red')
plot1.set(yscale="log")
plot2.set(yscale="log")
plt.title(f'BTC VS GOLD', fontsize = 16)
plt.ylabel("Closing price", size = 14)
plt.xlabel('Time', size = 14)
plt.legend(labels=["btc", "gold"],fontsize = 'large')
sns.set_style("dark")

plt.gca().xaxis.set_major_locator(mdates.DayLocator((1)))
plt.setp(plt.gca().get_xticklabels(), rotation=60, ha="right")
plt.tight_layout()


plt.show()
```
