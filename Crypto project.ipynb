# Euro, Pound and Yen currencies analysis and comparison to crypto currencies and gold

## Euro/dollar and pond/dollar historical price dataset

```python
eur_usd = pd.read_csv('euro-dollar-exchange-rate-historical-chart.csv', sep=";")
gbp_usd = pd.read_csv('pound-dollar-exchange-rate-historical-chart.csv', sep=";")
```

## Cleanning the dataset

```python
# deleting the first rows of the excel file
N = 8
eur_usd = eur_usd.iloc[N: , :]
gbp_usd = gbp_usd.iloc[N: , :]

#reseting indez
eur_usd = eur_usd.reset_index()
gbp_usd = gbp_usd.reset_index()

eur_usd = eur_usd.drop(['index'], axis = 1)
gbp_usd = gbp_usd.drop(['index'], axis = 1)

# separating columns
time_euro = eur_usd["Macrotrends Data Download"].apply(lambda x: x.split(','))
time_pound = gbp_usd["Macrotrends Data Download"].apply(lambda x: x.split(','))
```

```python
#cleanning columns of the dataframes so we can get a time column and a close price column

time_eur = list()
time_gbp = list()
euro = list()
pound = list()

for row in time_euro:
    time_eur.append(row[0])
    euro.append(row[1])

for row in time_pound:
    time_gbp.append(row[0])
    pound.append(row[1])
```

```python
eur_usd['time'] = time_eur
eur_usd['close_euro']= euro

gbp_usd['time'] = time_gbp
gbp_usd['close_pound']= pound

eur_usd = eur_usd.drop(['Macrotrends Data Download'], axis = 1)
gbp_usd = gbp_usd.drop(['Macrotrends Data Download'], axis = 1)

eur_usd['close_euro']= eur_usd['close_euro'].astype('float64')
eur_usd['time']= eur_usd['time'].astype('datetime64')

gbp_usd['close_pound']= gbp_usd['close_pound'].astype('float64')
gbp_usd['time']= gbp_usd['time'].astype('datetime64')
```
